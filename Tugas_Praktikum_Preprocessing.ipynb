**1. Memuat Data**

Import Library

```python
import pandas as pd
```

Data Movie dari github

Data frame

```python
movie_df = pd.read_csv('movie_sample_dataset.csv')
movie_df
```

**2. Memeriksa Data**

Memeriksa Tipe data dan Data yang Hilang

```python
movie_df.info()
```

```python
print(movie_df.isnull().sum())
```

**3. Membersihkan Data**

```python
movie_df.loc[:, 'color'] = movie_df['color'].str.strip().str.capitalize()
movie_df.loc[:, 'country'] = movie_df['country'].str.strip().str.upper()
movie_df = movie_df[(movie_df['duration']>=0)&
              (movie_df['gross']>=0)&
              (movie_df['title_year']>=0)&
              (movie_df['budget']>=0)&
              (movie_df['imdb_score']>=0)&
              (movie_df['movie_facebook_likes']>=0)]
movie_df = movie_df.replace('N/A', pd.NA)
movie_df.dropna(inplace=True)
movie_df
```

Dapat dilihat baris yang memiliki nilai NaN di kolom penting seperti gross dan budget sudah terhapus, nilai yang tidak konsisten atau kesalahan penulisan di kolom, seperti perbedaan antara "Color" dan "color" juga sudah diatasi, dan nilai-nilai yang tidak standar, seperti nilai negatif atau "N/A" sudah diubah dan dihapus. Hal itu dapat dilihat dengan kode dibawah ini

```python
print(movie_df.isnull().sum())
```

**4. Transformasi Data**

Cek Data

```python
movie_df.dtypes
```

Ubah Tipe Data

```python
movie_df['gross'] = movie_df['gross'].astype('int')
movie_df['budget'] = movie_df['budget'].astype('int')
movie_df.dtypes
```

Normalisasi Text

```python
movie_df['actors'] = movie_df['actors'].str.replace(',', ', ', regex=False)
movie_df['genres'] = movie_df['genres'].str.replace('|', ', ', regex=False)
movie_df
```

**5. Penyimpanan Data**

```python
movie_df.to_csv('movie_sample_dataset_cleaned.csv', index=False)
```
